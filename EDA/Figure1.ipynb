# Acc & Infer

```python
import matplotlib.pyplot as plt

# Data
models = [
    "HoViT-B", "HoViT-S", "ViT-Tiny", "LeViT-S", 
    "EfficientNet B0", "EfficientNet B1"
]
parameters = [8.3, 4.9, 5.5, 6.6, 4.0, 6.5]  # Parameters (M)
accuracy = [98.67, 98.63, 97.40, 95.87, 98.97, 98.54]  # Top-1 Accuracy (%)
latency = [9.57, 5.74, 10.01, 12.06, 10.49, 15.54]  # Inference Time (ms)
colors = ['green', 'orange', 'blue', 'blue', 'blue', 'blue', 'red']  # Colors for the models

# First graph: Accuracy vs Inference Time
plt.figure(figsize=(10, 6))
for i in range(len(models)):
    plt.scatter(latency[i], accuracy[i], label=models[i], color=colors[i], s=100)

for i, model in enumerate(models):
    plt.text(latency[i] + 0.5, accuracy[i], model, fontsize=10)

x_arange = [0, 400]

# plt.xscale('log')
plt.xlim(x_arange)
# log scale
# plt.xscale('log')
plt.xlabel("Inference Time (ms)", fontsize=12)
plt.ylabel("Top-1 Accuracy (%)", fontsize=12)
plt.title("Accuracy vs Inference Time", fontsize=14)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend(loc="lower left", fontsize=10, bbox_to_anchor=(1, 0))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = [
    "HoViT (Ours)", "ViT_patch8", "LeViT_128", "ResNet50", 
    "EfficientNet-b0", "EfficientNet-b1"
]
parameters = [4.9, 85.8, 8.4, 23.5, 4.0, 6.5]  # Parameters (M)
accuracy = [99.03, 93.36, 97.67, 98.75, 98.88, 98.76]  # Top-1 Accuracy (%)
latency = [5.74, 282.60, 14.32, 13.29, 12.79, 14.06]  # Inference Time (ms)
colors = ['green', 'orange', 'blue', 'blue', 'blue', 'blue']  # Colors for the models

# Custom scale function (Log for 0-200, Linear for 200+)
def custom_xscale(x):
    if x <= 200:
        return np.log10(x + 1)  # Apply log scale for 0-200
    else:
        return 2 + (x - 200) / 100  # Linear scaling for 200+

# Apply custom scale to latency
scaled_latency = [custom_xscale(x) for x in latency]

# First graph: Accuracy vs Inference Time with custom scale
plt.figure(figsize=(10, 6))
for i in range(len(models)):
    plt.scatter(scaled_latency[i], accuracy[i], label=models[i], color=colors[i], s=100)

for i, model in enumerate(models):
    plt.text(scaled_latency[i], accuracy[i], model, fontsize=10)

# Custom ticks for x-axis
ticks = [1, 10, 50, 100, 200, 300, 400]
scaled_ticks = [custom_xscale(t) for t in ticks]
plt.xticks(scaled_ticks, labels=ticks)

plt.xlabel("Inference Time (ms)", fontsize=12)
plt.ylabel("Top-1 Accuracy (%)", fontsize=12)
plt.title("Accuracy vs Custom Scaled Inference Time", fontsize=14)
plt.grid(True, linestyle="--", linewidth=0.5)
plt.legend(loc="lower left", fontsize=10, bbox_to_anchor=(1, 0))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = [
    "HoViT (Ours)", "ViT_patch8", "LeViT_128", "ResNet50", 
    "EfficientNet-b0", "EfficientNet-b1"
]
parameters = [4.9, 85.8, 8.4, 23.5, 4.0, 6.5]  # Parameters (M)
accuracy = [99.03, 93.36, 97.67, 98.75, 98.88, 98.76]  # Top-1 Accuracy (%)
latency = [5.74, 282.60, 14.32, 13.29, 12.79, 14.06]  # Inference Time (ms)
colors = ['green', 'red', 'purple', 'orange', 'blue', 'blue']  # Colors for the models

# First graph: Accuracy vs Inference Time
plt.figure(figsize=(10, 6))
for i in range(len(models)):
    plt.scatter(latency[i], accuracy[i], label=models[i], color=colors[i], s=100)

    # Annotate with adjusted text positions and arrows
    x_offset = -60 if models[i] in ["ResNet50", "ViT_patch8"] else 60
    y_offset = -20 if models[i] in [ "EfficientNet-b1"] else 0
    plt.annotate(
        models[i],
        (latency[i], accuracy[i]),
        textcoords="offset points",
        xytext=(x_offset, y_offset),
        ha="center",
        arrowprops=dict(arrowstyle="->", color=colors[i], lw=1)
    )

x_ticks = [5, 10, 100, 200, 300]

plt.xscale('log')
plt.xlabel("Inference Time (ms)", fontsize=12)
plt.ylabel("Top-1 Accuracy (%)", fontsize=12)
plt.xticks(x_ticks, labels=x_ticks)
# plt.title("Accuracy vs Inference Time", fontsize=14)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend(loc="lower left", fontsize=10, bbox_to_anchor=(1, 0))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
models = [
    "HoViT-B", "HoViT-S", "ViT-Tiny", "LeViT-S", 
    "EfficientNet B0", "EfficientNet B1"
]
parameters = [8.3, 4.9, 5.5, 6.6, 4.0, 6.5]  # Parameters (M)
accuracy = [98.67, 98.63, 97.40, 95.87, 98.97, 98.54]  # Top-1 Accuracy (%)
latency = [9.57, 5.74, 10.01, 12.06, 10.49, 15.54]  # Inference Time (ms)
colors = ['orange', 'orange', 'blue', 'blue', 'green', 'green']  # Colors for the models

# Create a figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6), gridspec_kw={'width_ratios': [1, 1]})

# First graph: Accuracy vs Inference Time
for i in range(len(models)):
    axes[0].scatter(latency[i], accuracy[i], label=models[i], color=colors[i], s=100)
    x_offset = -60 if models[i] in ["EfficientNet B1", "ViT_patch8"] else 60
    y_offset = -2 if models[i] in ["EfficientNet B1"] else 0
    axes[0].annotate(
        models[i],
        (latency[i], accuracy[i]),
        textcoords="offset points",
        xytext=(x_offset, y_offset),
        ha="center",
        arrowprops=dict(arrowstyle="->", color=colors[i], lw=1)
    )
# axes[0].set_xscale('log')
axes[0].set_xlabel("Inference Time (ms)", fontsize=12)
axes[0].set_ylabel("Top-1 Accuracy (%)", fontsize=12)
axes[0].set_xticks([0, 2, 4, 6, 8, 10, 12, 14, 16])
axes[0].set_xticklabels([0, 2, 4, 6, 8, 10, 12, 14, 16])
axes[0].grid(True, which="both", linestyle="--", linewidth=0.5)
axes[0].set_title("Accuracy vs Inference Time", fontsize=14)

# Second graph: Accuracy vs Parameters
for i in range(len(models)):
    axes[1].scatter(parameters[i], accuracy[i], label=models[i], color=colors[i], s=100)
    x_offset = -60 if models[i] in ["ViT_patch8"] else 60
    y_offset = -5 if models[i] in ["EfficientNet B1"] else 0
    if models[i] == "EfficientNet-b0":
        y_offset = -5
    axes[1].annotate(
        models[i],
        (parameters[i], accuracy[i]),
        textcoords="offset points",
        xytext=(x_offset, y_offset),
        ha="center",
        arrowprops=dict(arrowstyle="->", color=colors[i], lw=1)
    )
axes[1].set_xticks([0, 2, 4, 6, 8, 10])
axes[1].set_xticklabels([0, 2, 4, 6, 8, 10])
axes[1].set_xlabel("Parameters (M)", fontsize=12)
axes[1].set_ylabel("Top-1 Accuracy (%)", fontsize=12)
axes[1].grid(True, which="both", linestyle="--", linewidth=0.5)
axes[1].set_title("Accuracy vs Parameters", fontsize=14)

# Common legend
fig.legend(models, loc="upper center", ncol=6, fontsize=10)
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to fit the legend
plt.show()
```
